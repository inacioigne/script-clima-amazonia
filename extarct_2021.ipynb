```python
from os import listdir
import pymupdf
from cloudinary import uploader, config, api
import json
```

```python
path = 'data/pdf/2021'
pdfs = sorted(listdir(path))
len(pdfs)
```

```python
for pdf in pdfs:
    mmdd = pdf[8:12]
    pdf_path = f'{path}/{pdf}'
    doc = pymupdf.open(pdf_path)
    page = doc.load_page(0)
    pix = page.get_pixmap()
    pix.save(f"test/{mmdd}.png")
    print(mmdd)
```

```python
img_path = 'test'
issues = []
for img in listdir(img_path):
    name = img.split('.')[0]
    img_full_path = f'{img_path}/{img}'
    result = uploader.upload(
                img_full_path,
                public_id=name,
                folder=f"clima-amazonia/2021",
                use_filename=True,
                overwrite=True,
                unique_filename=True
            )
    url = result['secure_url']
    ed = {'number': name, 'url': "", "img": url}
    issues.append(ed)
    print(name, url)
```

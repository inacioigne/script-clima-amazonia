# Boletins

```python
path_boletins = Path('/home/inacio/clima-amazonia/data/boletim/2026')
boletins = listdir(path_boletins)
for b in boletins:
    path_boletim = Path(path_boletins / b)
    with open(path_boletim, 'r') as f:
        data = json.load(f)
    for section in data['images'].keys():
        for k in data['images'][section].keys():
            url = data['images'][section][k]
            if 'clima-amazonia' in url:
                x = url.split('clima-amazonia')[1]
            if 'libris.shop' in url:
                x = url.split('libris.shop')[1]
                
            new_url = f'https://climamazonia.shop{x}'.replace('fev/', '')
            if 'mar/' in new_url:
                new_url = new_url.replace('mar/', '')
            
            
            new_url = new_url.split('.')[:-1]
            new_url = '.'.join(new_url) + '.webp'
            data['images'][section][k] = new_url
    with open(path_boletim, 'w') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
            
    print(path_boletim)
            
```
